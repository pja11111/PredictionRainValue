```python
#ตรวจสอบค่า Outlier ใน Dataset
df = pd.read_csv('dry_wet_datacleaned.csv')

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df < lower_bound) | (df > upper_bound))

print(outliers.sum())
```

```python
#กำจัดค่า Outlier ด้วยค่าเฉลี่ยของแต่ละคอลลัมน์และแสดงผลค่าที่กำจัดแล้ว
df = pd.read_csv('dry_wet_datacleaned.csv')

columns_to_remove_outliers = [ 'DSL (Day)', 'MAX', 'MEAN', 'MIN', 'R35mm (Day)']

Q1 = df[columns_to_remove_outliers].quantile(0.25)
Q3 = df[columns_to_remove_outliers].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[columns_to_remove_outliers] < lower_bound) | (df[columns_to_remove_outliers] > upper_bound))

column_means = df[columns_to_remove_outliers].mean()

for column in columns_to_remove_outliers:
    df[column] = df[column].apply(lambda x: column_means[column] if x in outliers[column] else x)

df.to_csv('dry_wet_datacleaned.csv', index=False)
df = pd.read_csv('dry_wet_datacleaned.csv')
pd.DataFrame(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
##from sklearn.neural_network import MLPRegressor
##from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python
# แบ่งข้อมูลเป็นชุดการฝึกและชุดทดสอบ
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
def calculate_metrics(X, Y, model):
    pred = model.predict(X)
    mse = mean_squared_error(Y, pred)
    rmse = mse ** 0.5
    mae = mean_absolute_error(Y, pred)
    r2 = r2_score(Y, pred)
    return rmse, mae, r2
```

```python
# สร้างโมเดล Linear Regression
lr = LinearRegression()

# ฝึกโมเดลโดยใช้ชุดการฝึก
lr.fit(X_train, Y_train)
```

```python
svr = SVR(kernel='linear')
svr.fit(X_train, Y_train)
```

```python

# สร้างโมเดล Random Forest
random_forest_model = RandomForestRegressor()

# ฝึกโมเดลโดยใช้ชุดฝึก
random_forest_model.fit(X_train, Y_train)
```

```python
# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
linear_y_pred = linear_model.predict(X_test[:30])

# SVR
svr_model = SVR()
svr_model.fit(X_train, Y_train)
svr_y_pred = svr_model.predict(X_test[:30])

# Calculate MSE and R2 for Linear Regression
linear_mse = mean_squared_error(Y_test[:30], linear_y_pred)
linear_r2 = r2_score(Y_test[:30], linear_y_pred)

# Calculate MSE and R2 for SVR
svr_mse = mean_squared_error(Y_test[:30], svr_y_pred)
svr_r2 = r2_score(Y_test[:30], svr_y_pred)

# Compare the results
print("Linear Regression:")
print(f'Mean Squared Error (MSE): {linear_mse}')
print(f'R-squared (R2): {linear_r2}')

print("SVR:")
print(f'Mean Squared Error (MSE): {svr_mse}')
print(f'R-squared (R2): {svr_r2}')

models = ['Linear Regression', 'SVR']
mse_values = [linear_mse, svr_mse]

plt.bar(models, mse_values)
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Comparison of Mean Squared Error (MSE) between Models')
plt.show()

# Create a bar chart to compare R2
r2_values = [linear_r2, svr_r2]

plt.bar(models, r2_values)
plt.ylabel('R-squared (R2)')
plt.title('Comparison of R-squared (R2) between Models')
plt.show()
```

```python
# Random Forest
random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train, Y_train)
random_forest_y_pred = random_forest_model.predict(X_test[:30])

# Calculate MSE and R2 for Random Forest
random_forest_mse = mean_squared_error(Y_test[:30], random_forest_y_pred)
random_forest_r2 = r2_score(Y_test[:30], random_forest_y_pred)

# Compare the results
print("Random Forest:")
print(f'Mean Squared Error (MSE): {random_forest_mse}')
print(f'R-squared (R2): {random_forest_r2}')

models.append('Random Forest')
mse_values.append(random_forest_mse)
r2_values.append(random_forest_r2)

# Create a bar chart to compare MSE
plt.bar(models, mse_values)
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Comparison of Mean Squared Error (MSE) between Models')
plt.show()

# Create a bar chart to compare R2
plt.bar(models, r2_values)
plt.ylabel('R-squared (R2)')
plt.title('Comparison of R-squared (R2) between Models')
plt.show()
```
